## Load Datasets

Data description:
1. NY_GDP_MKTP_KD_ZG:  [Annual GDP growth (%)](https://unstats.un.org/sdgs/dataportal/database)
2. SI_POV_NAHC: Poverty rate [(Proportion of population living below the national poverty line (%))](https://unstats.un.org/sdgs/dataportal/database)
3. P_Data_Extract_From_World_Development_Indicators (4): Employment rate [(Employers, total (% of total employment) (modeled ILO estimate))](https://databank.worldbank.org/source/world-development-indicators)
4. P_Data_Extract_From_World_Development_Indicators (5): [Foreign Direct Investment (FDI), net inflows (% of GDP)](https://databank.worldbank.org/source/world-development-indicators)
5. P_Data_Extract_From_World_Development_Indicators (6): [Foreign Direct Investment (FDI), net outflows (% of GDP)](https://databank.worldbank.org/source/world-development-indicators)

```python
SHEET_NAME = 'Table format', 'Data'
gdp_growth_original = pd.read_excel('Datasets/NY_GDP_MKTP_KD_ZG.xlsx', sheet_name=SHEET_NAME[0])
poverty_rate_original = pd.read_excel('Datasets/SI_POV_NAHC.xlsx', sheet_name=SHEET_NAME[0])
employment_rate_original = pd.read_excel('Datasets/P_Data_Extract_From_World_Development_Indicators (4).xlsx', sheet_name=SHEET_NAME[1])
fdi_net_inflows_original = pd.read_excel('Datasets/P_Data_Extract_From_World_Development_Indicators (5).xlsx', sheet_name=SHEET_NAME[1])
fdi_net_outflows_original = pd.read_excel('Datasets/P_Data_Extract_From_World_Development_Indicators (6).xlsx', sheet_name=SHEET_NAME[1])
```

## Data Wrangling (Data Preprocessing and Data Transformation)

### gdp_growth

```python
gdp_growth_df = gdp_growth_original.copy()

# Drop unnecessary column
gdp_growth_df.drop(columns=['Goal', 'Target', 'Indicator', 'SeriesCode', 'SeriesDescription', 'GeoAreaCode', 'Reporting Type', 'Units'], inplace=True)

# Rename column
gdp_growth_df.rename(columns={'GeoAreaName': 'Country Name'}, inplace=True)
gdp_growth_df.columns = gdp_growth_df.columns.str.strip()

# Missing value handling
gdp_growth_df[gdp_growth_df.columns.difference(['Country Name'])] = gdp_growth_df[gdp_growth_df.columns.difference(['Country Name'])].fillna(method='ffill', axis=1)
gdp_growth_df[gdp_growth_df.columns.difference(['Country Name'])] = gdp_growth_df[gdp_growth_df.columns.difference(['Country Name'])].fillna(method='bfill', axis=1)

# Change index
gdp_growth_df.set_index('Country Name', inplace=True)
```

```python
gdp_growth_df.head(3)
```

### poverty_rate

```python
poverty_rate_df = poverty_rate_original.copy()

# Remove the record / data with the urban location
poverty_rate_df = poverty_rate_df[poverty_rate_df['Location'] == 'ALLAREA']

# Drop / remove unnecessary columns
poverty_rate_df.drop(columns=['Goal', 'Target', 'Indicator', 'SeriesCode', 'SeriesDescription', 'GeoAreaCode', 'Reporting Type', 'Units', 'Age', 'Sex', '2023', 'Location'], inplace=True)

# Rename column
poverty_rate_df.rename(columns={'GeoAreaName': 'Country Name'}, inplace=True)
poverty_rate_df.columns = poverty_rate_df.columns.str.strip()

# Missing values handling
poverty_rate_df[poverty_rate_df.columns.difference(['Country Name'])] = poverty_rate_df[poverty_rate_df.columns.difference(['Country Name'])].fillna(method='ffill', axis=1)
poverty_rate_df[poverty_rate_df.columns.difference(['Country Name'])] = poverty_rate_df[poverty_rate_df.columns.difference(['Country Name'])].fillna(method='bfill', axis=1)

# Change index
poverty_rate_df.set_index('Country Name', inplace=True)
```

```python
poverty_rate_df.head(3)
```

### employment_rate

```python
employment_rate_df = employment_rate_original.copy()

# Drop / remove unnecessary columns
employment_rate_df.drop(columns=['Series Name', 'Series Code', 'Country Code', '2023 [YR2023]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]'], inplace=True)

# Rename columns
pattern = r"\[.*?\]"
employment_rate_df.columns = map(lambda x: re.sub(pattern, '', x),employment_rate_df.columns)
employment_rate_df.columns = employment_rate_df.columns.str.strip()

# Missing values handling
employment_rate_df.replace(to_replace='..', value=np.NaN, inplace=True)
employment_rate_df[employment_rate_df.columns.difference(['Country Name'])] = employment_rate_df[employment_rate_df.columns.difference(['Country Name'])].fillna(method='ffill', axis=1)
employment_rate_df[employment_rate_df.columns.difference(['Country Name'])] = employment_rate_df[employment_rate_df.columns.difference(['Country Name'])].fillna(method='bfill', axis=1)
employment_rate_df.dropna(inplace=True)

# Change index
employment_rate_df.set_index('Country Name', inplace=True)
```

```python
employment_rate_df.head(3)
```

### fdi_net_inflows

```python
fdi_net_inflows_df = fdi_net_inflows_original.copy()

# Drop / remove unnecessary columns
fdi_net_inflows_df.drop(columns=['Series Name', 'Series Code', 'Country Code', '2023 [YR2023]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]'], inplace=True)

# Rename columns
pattern = r"\[.*?\]"
fdi_net_inflows_df.columns = map(lambda x: re.sub(pattern, '', x),fdi_net_inflows_df.columns)
fdi_net_inflows_df.columns = fdi_net_inflows_df.columns.str.strip()

# Missing values handling
fdi_net_inflows_df.replace(to_replace='..', value=np.NaN, inplace=True)
fdi_net_inflows_df[fdi_net_inflows_df.columns.difference(['Country Name'])] = fdi_net_inflows_df[fdi_net_inflows_df.columns.difference(['Country Name'])].fillna(method='ffill', axis=1)
fdi_net_inflows_df[fdi_net_inflows_df.columns.difference(['Country Name'])] = fdi_net_inflows_df[fdi_net_inflows_df.columns.difference(['Country Name'])].fillna(method='bfill', axis=1)
fdi_net_inflows_df.dropna(inplace=True)

# Change index
fdi_net_inflows_df.set_index('Country Name', inplace=True)
```

```python
fdi_net_inflows_df.head(3)
```

### fdi_net_outflows

```python
fdi_net_outflows_df = fdi_net_outflows_original.copy()

# Drop / remove unnecessary columns
fdi_net_outflows_df.drop(columns=['Series Name', 'Series Code', 'Country Code', '2023 [YR2023]', '2009 [YR2009]', '2010 [YR2010]', '2011 [YR2011]'], inplace=True)

# Rename columns
pattern = r"\[.*?\]"
fdi_net_outflows_df.columns = map(lambda x: re.sub(pattern, '', x),fdi_net_outflows_df.columns)
fdi_net_outflows_df.columns = fdi_net_outflows_df.columns.str.strip()


# Missing values handling
fdi_net_outflows_df.replace(to_replace='..', value=np.NaN, inplace=True)
fdi_net_outflows_df[fdi_net_outflows_df.columns.difference(['Country Name'])] = fdi_net_outflows_df[fdi_net_outflows_df.columns.difference(['Country Name'])].fillna(method='ffill', axis=1)
fdi_net_outflows_df[fdi_net_outflows_df.columns.difference(['Country Name'])] = fdi_net_outflows_df[fdi_net_outflows_df.columns.difference(['Country Name'])].fillna(method='bfill', axis=1)
fdi_net_outflows_df.dropna(inplace=True)

# Change index
fdi_net_outflows_df.set_index('Country Name', inplace=True)
```

```python
fdi_net_outflows_df.head(3)
```

## Data Integration

```python
gdp_growth_df = pd.DataFrame(gdp_growth_df.stack().reset_index()).rename(columns={'level_1': 'Year', 0:'GDP Growth'})
poverty_rate_df = pd.DataFrame(poverty_rate_df.stack().reset_index()).rename(columns={'level_1': 'Year', 0:'Poverty Rate'})
employment_rate_df = pd.DataFrame(employment_rate_df.stack().reset_index()).rename(columns={'level_1': 'Year', 0:'Employment Rate'})
fdi_net_inflows_df = pd.DataFrame(fdi_net_inflows_df.stack().reset_index()).rename(columns={'level_1': 'Year', 0:'FDI Net Inflows'})
fdi_net_outflows_df = pd.DataFrame(fdi_net_outflows_df.stack().reset_index()).rename(columns={'level_1': 'Year', 0:'FDI Net Outflows'})
```
